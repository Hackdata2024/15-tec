```python
df_new = pd.read_csv('precipitate.csv')

df_flood = pd.read_csv("final.csv")
df_ppt = pd.read_csv('precipitate.csv')

df_flood_2d = df_flood.to_numpy().tolist()
df_ppt_2d = df_ppt.to_numpy().tolist()

new_df_flood_2d = []

def avg(list):
    sum = 0
    for i in list:
        sum += i
    return int(sum)

for record in df_flood_2d:
    state = record[2]
    month = record[1].split('-')[0]
    year = record[1].split('-')[1]

    if not month.isnumeric() or not year.isnumeric():
        continue

    for record2 in df_ppt_2d:
        if not str(record2[1]).isnumeric():
            continue
        
        if state in record2[0].lower().replace("&", "and") and year == str(record2[1]):
            record.append(record2[int(month) + 1])
            new_df_flood_2d.append(record)

i = 0
while (i < len(new_df_flood_2d)):
    record = new_df_flood_2d[i]
    avg_of_ppt = record[slice(3, len(record))]
    record = [record[0], record[1], record[2], avg(avg_of_ppt)]
    new_df_flood_2d[i] = record
    i += 1

pd.DataFrame(np.array(new_df_flood_2d)).to_csv("flood_with_ppt.csv")
            
```
